```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/shamswardak/INST447-Final-Project/main/virginia_county_data.csv'
vacounty_df = pd.read_csv(url)
```

```python
vacounty_df.head()
```

In this part of the program, we created a database from the dataset that we found from the Virginia website however, we implemented it on a github so it could have easier access when it comes to implementing the dataset. And got a fair amount of columns in the dataset that we can manipulate.

```python
vacounty_df.info()
```

Right here we are starting to prepare to clean the dataframe and look for anything that seems out of place and make sure that the data type for each portion is something that we work with if not then we would have to change the data type accordingly.

```python
vacounty_df = vacounty_df.drop(['fips', 'stateCode', 'state', 'growthSince2010', 'name', 'slug','density', 'StatePercentage', 'StateRank', 'area'], axis=1)
vacounty_df
```

What we did right after was remove the columns that we found were similar than other columns and removed columns that we feel we wouldn't find useful for the assignment and save it to the same dataframe to make sure that we did the according changes right.

```python
vacounty_df = vacounty_df.sort_values('pop2023', ascending=True)
vacounty_df
```

For this portion of the program I decided to resort the values to see which county has the highest population count and wich once has the lowest. This will also help out when it comes to graphing the data

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.scatter(vacounty_df.ctyname, vacounty_df.pop2023)

ax.set_xlabel('VA County')
ax.set_ylabel('Population')
ax.set_title('Total amount of population per VA county')

plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

When it came to graphing the dataframe I grabbed my x and y value accordingly, Then I created a fig and ax so I can set the x and y labels and also change the figure size so it wouldn't look squished and instead make it seem more spread out. Then I rotated the xtick so all the x titles will be more readable and tighten the layout to make the whole x value titles more oraganized. Based on the result that I got in this graph it shows that the highest county with the most population is going to be Fairfax County followed by Prince Williams, Virginia beach, and Loudoun Chesterfield County

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(vacounty_df.ctyname, vacounty_df.landArea, marker='o', linestyle='-', color='blue')

ax.set_xlabel('VA County')
ax.set_ylabel('Land Area')
ax.set_title('Total amount of land in the VA county')

plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

For this portion I did very similar towards the graph above but the difference I did was make it a regular plot and set the according values such as marker, linestyle, and color so that it would look more readable and appealing. Based on the results on this graph it shows that there are a few counties that have a high number of land area and a large amount of counties which land area are very small.
